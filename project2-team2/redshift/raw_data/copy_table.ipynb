# copy table

s3에 매일 올라오는 csv 파일을 COPY로 복사

- movie_codes
- movie_company
- movie_director
- movie_genre
- movie_grade
- movie_sale
- movie_summary

```redshift
COPY raw_data.movie_codes
FROM 's3://your_bucket/your_file.csv'
credentials 'aws_iam_role=your_redshift_arn'
delimiter ',' dateformat 'auto' timeformat 'auto' IGNOREHEADER 1 removequotes;
```
