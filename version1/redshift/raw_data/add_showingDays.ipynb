# showingDays 추가

- movie_sale 테이블에 showingDays를 추가
- 영화 개봉일부터 date까지 며칠동안 상영하고 있는지 확인할 수 있는 컬럼
- 영화 개봉일이 date보다 더 늦은 경우, 해당 컬럼 제거 (showingDays 계산 참고)

### add column

```redshift
ALTER TABLE raw_data.movie_sale
ADD COLUMN showingDays FLOAT;
```

### showingDays 계산

```redshift
BEGIN;

UPDATE raw_data.movie_sale
SET showingDays = DATEDIFF('day', openDt, date) + 1
WHERE DATEDIFF('day', openDt, date) > 0;

DELETE FROM raw_data.movie_sale
WHERE DATEDIFF('day', openDt, date) <= 0 OR showingDays IS NULL;

COMMIT;
```
