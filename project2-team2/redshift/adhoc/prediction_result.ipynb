# SageMaker로 얻은 예측 정보를 저장하는 테이블

### prediction_result 테이블 생성

```redshift
CREATE TABLE adhoc.prediction_result(
    serialNum INTEGER,
    date DATE,
    movieCd INTEGER primary key,
    salesAmt FLOAT,
    genres VARCHAR(100),
    ticketRatio FLOAT,
    openDt DATE,
    showingDays FLOAT,
    predicted_sales FLOAT
);
```

### s3에 있는 prediction_result.csv COPY

```redshift
COPY adhoc.prediction_result
FROM 's3://your_bucket/your_file.csv'
credentials 'aws_iam_role=your_redshift_arn'
delimiter ',' dateformat 'auto' timeformat 'auto' IGNOREHEADER 1 removequotes;
```

```redshift
select * from adhoc.prediction_result
```
